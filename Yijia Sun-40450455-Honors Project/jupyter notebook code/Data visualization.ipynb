```python
import matplotlib.pyplot as plt
import pandas as pd
java_data= pd.read_excel('C:/Users/HP/Desktop/honors project/cleaned data/51java-xlsx-3-xlsx.xlsx')
```

```python
import numpy as np
```

```python
fig = plt.figure(figsize=(28,21))
```

```python
data_count=np.arange(14)
data_city=[]
for i in range(8):
    data_count[i]=java_data.joblocation_detail[java_data.experience_name=='Fresh graduate'].value_counts()[i]
    data_city.append(java_data.joblocation_detail[java_data.experience_name=='Fresh graduate'].value_counts().index[i])
for i in range(8,14):
    data_count[i]=java_data.joblocation_detail[java_data.experience_name=='Fresh graduate'].value_counts()[i+1]
    data_city.append(java_data.joblocation_detail[java_data.experience_name=='Fresh graduate'].value_counts().index[i+1])
```

```python
data_city
#data_city.append(data_city1)
#data_city
```

```python
from matplotlib import cm
```

```python

color = cm.jet(np.array(data_count)/max(data_count))
plt.bar( data_city,data_count,width=0.5,color=color)
plt.xticks(data_city,data_city, rotation=90)
plt.title('The relationship between the number of Posts and the City(Fresh graduate)')
for a, b in zip(data_city,data_count):
        plt.text(a, b + 0.05, '%.0f' % b, ha='center', va='bottom', fontsize=9) 
plt.xlabel('city')
plt.ylabel('count')
#plt.savefig('C:/Users/HP/Desktop/honors project/cleaned data/Posts and the City.jpg')
plt.show()
```
