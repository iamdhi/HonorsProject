```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
java_data= pd.read_excel('C:/Users/HP/Desktop/honors project/cleaned data/HOT INDUSTRY.xlsx')
```

```python
fig = plt.figure(figsize=(28,21))
```

```python
group_location = java_data.groupby('job_industry_name')
avg_salary_min = group_location.aggregate({'salary_min_BGP':np.mean}).sort_values('salary_min_BGP', ascending = False)
avg_salary_max = group_location.aggregate({'salary_max_BGP':np.mean}).sort_values('salary_max_BGP', ascending = False)
print(avg_salary_max)
print(avg_salary_min)
```

```python
region_industry = []
for region in avg_salary_max.index:
    region_industry.append(java_data.salary_max_BGP[java_data.job_industry_name == region])
    
plt.boxplot(x = region_industry, 
             patch_artist=True,
             labels = avg_salary_max.index, # 添加x轴的刻度标签
             showmeans=True, 
             #boxprops = {'color':'black', 'facecolor':'steelblue'}, 
             #flierprops = {'marker':'o','markerfacecolor':'red', 'markersize':3}, 
             meanprops = {'marker':'D','markerfacecolor':'indianred', 'markersize':4},
             medianprops = {'linestyle':'--','color':'orange'}
            )
plt.title('The relationship between salary_max and Industry')
plt.ylabel('salary_max')
plt.xlabel('Industry')
plt.xticks(rotation=90)
plt.show
```
